```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df= df.astype({'Gender': 'bool'})
df.dtypes
```

```python
fig = plt.figure(figsize = (15,20))
ax = fig.gca()
df.hist(ax = ax)
```

```python
fig = plt.figure(figsize = (10,5))
ax = fig.gca()
df["Area (ft.)"].hist(ax=ax, bins=7)
```

```python
fig = plt.figure(figsize = (10,5))
ax = fig.gca()
df["Price"].hist(ax=ax, bins=7)
```

```python
x = df[df["Mortgage"]== 'No']["Price"]
y = df[df["Mortgage"]== 'Yes']["Price"]

bins = 20

plt.hist(x, bins=bins, alpha=0.5, label='Without Mortgage')
plt.hist(y, bins=bins, alpha=0.5, label='With Mortgage')
plt.legend(loc='upper right')
plt.show()
```

```python
x = df[df["Building"]== 1]["Price"]
y = df[df["Building"]== 5]["Price"]

bins = 20

plt.hist(x, bins=bins, alpha=0.5, label='Price of Building 1')
plt.hist(y, bins=bins, alpha=0.5, label='Price of Building 5')
plt.legend(loc='upper right')
plt.show()
```

```python
x = df[df["Purpose"]== 'Home']["Price"]
y = df[df["Purpose"]== 'Investment']["Price"]

bins = 20

plt.hist(x, bins=bins, alpha=0.5, label='Home')
plt.hist(y, bins=bins, alpha=0.5, label='Investment')
plt.legend(loc='upper right')
plt.show()
```

```python
x = df[df["Status"]== "Sold"]["Price"]
print(x.shape)
y = df[df["Status"].isnull()]["Price"]
print(y.shape)
```

```python
x = df[df["Status"]== "Sold"]["Price"]
y = df[df["Status"].isnull()]["Price"]

bins = 20

plt.hist(x, bins=bins, alpha=0.5, label='Price of sold properties')
plt.hist(y, bins=bins, alpha=0.5, label='Price of unsold properties')
plt.legend(loc='upper right')
plt.show()
```

```python
ax = df.plot.scatter(x='Price', y='Area (ft.)', figsize=(8, 4), marker='$\u25EF$')
ax.set_xlabel('Price')
ax.set_ylabel('Area')
ax.axhline(0, color='grey', lw=1)
ax.axvline(0, color='grey', lw=1)

plt.tight_layout()
plt.show()
```

```python
from sklearn.utils import resample
results = []
for nrepeat in range(1000):
    sample = resample(df['Price'])
    results.append(sample.median())
results = pd.Series(results)
fig = plt.figure(figsize = (5,3))
ax = fig.gca()
results.hist(ax = ax)
```
